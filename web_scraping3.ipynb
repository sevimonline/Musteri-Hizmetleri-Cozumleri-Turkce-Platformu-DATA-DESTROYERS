```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as bts
from urllib.parse import urljoin  # urljoin'u import edin
import time
from selenium.webdriver.common.action_chains import ActionChains

def getAndParseURL(url):
    driver.get(url)
    time.sleep(2)
    html = driver.page_source
    soup = bts(html, 'html.parser')
    return soup

# WebDriver'ı başlat
driver = webdriver.Chrome()

base_url = 'https://www.sikayetvar.com/turkcell/superbox'

# Şikayet sayfalarını dolaşarak ürün linklerini çek
pages = []
for page in range(, 38):  # Örnek olarak 1'den 2'ye kadar olan sayfaları ziyaret ediyoruz
    pages.append(f"{base_url}?page={page}")

# Başlıkları ve linklerini saklamak için bir liste oluşturun
products = []

for page in pages:
    soup = getAndParseURL(page)
    titles = soup.find_all('article', class_='card-v2 card-v3 ga-v ga-c')
    for title in titles:
        a_tag = title.find('a')
        if a_tag is not None:
            link = a_tag['href']
            full_link = urljoin(base_url, link)
            products.append(full_link)

# Şikayet detaylarını almak için fonksiyon
def getProductDetails(product_url):
    # Web sayfasını aç
    driver.get(product_url)
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'h1.complaint-detail-title')))
    time.sleep(2)
    
    product_name = driver.find_element(By.CSS_SELECTOR, 'h1.complaint-detail-title').text
    comment_paragraphs = driver.find_elements(By.CSS_SELECTOR, 'div.complaint-detail-description')
    comments = [paragraph.text for paragraph in comment_paragraphs]

    # Her bir ürün için bir sözlük oluştur
    product_dict = {
        'yorum': comments,
        'product_name': product_name
    }

    return product_dict

all_yorumlar = []

for product in products:
    details = getProductDetails(product)
    all_yorumlar.append(details)

# Tarayıcıyı kapatma
driver.quit()

# Sonuçları yazdırma
for yorum in all_yorumlar:
    print(yorum)
```

```python
import pandas as pd
# Create a DataFrame from the collected data
df = pd.DataFrame(all_yorumlar)
df.to_csv('---.csv', index=False)
df
```
