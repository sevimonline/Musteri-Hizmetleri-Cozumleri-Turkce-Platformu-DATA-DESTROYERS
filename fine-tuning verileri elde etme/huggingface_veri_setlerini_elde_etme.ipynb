```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
import os
from datasets import load_dataset
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, "html.parser")
    return soup
```

```python
pages = []
for page in range(0,8):
    pages.append("https://huggingface.co/datasets?modality=modality:text&p="+str(page)+"&sort=trending&search=turkish")
    
pages
```

```python
links = []

for page in pages:
    html = getAndParseURL(page)
    for product in html.findAll("div", {"class": "w-full truncate"}):
        h4_tag = product.find("h4", {"class": "text-md truncate font-mono text-black dark:group-hover/repo:text-yellow-500 group-hover/repo:text-red-600 text-smd"})
        if h4_tag:
            links.append("https://huggingface.co/datasets/" + h4_tag.text.strip())
```

```python
links
```

```python
def getAndParseURL(url):
    response = requests.get(url)
    return BeautifulSoup(response.text, 'html.parser')

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--version=Chrome/119.0.6045.105")

chromedriver_path = "driver_folder/chromedriver/chromedriver.exe"
driver = webdriver.Chrome(options=chrome_options)

data = []

for link in links:
    driver.get(link)
    
    try:
        use_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(@class, "btn w-full cursor-pointer text-sm")]'))
        )
        use_button.click()
        

        datasets_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(@class, "flex w-full cursor-pointer items-center whitespace-nowrap")]'))
        )
        datasets_button.click()
        
        
        code_block = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//pre[contains(@class, "overflow-x-auto px-4 py-5 text-sm text-gray-700 sm:px-6")]'))
        )
        code_text = code_block.text
        
        
        data.append(code_text)
        
    except Exception as e:
        print(f"An error occurred: {e}")
        continue

driver.quit()


df = pd.DataFrame(data, columns=['Code'])
df.to_csv('extracted_code.csv', index=False)
print(df)
```
